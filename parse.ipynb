```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
x = [] 
y = [] 
  
with open('piechart.csv','r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
      
    for row in plots: 
        x.append(row[0]) 
        y.append(row[1])
x.pop(0)
y.pop(0)
```

```python
labels = x
sizes = y

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
x = [] 
y = [] 
graph = []
  
with open('ukdata.csv','r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    print(plots)
    for row in plots: 
        graph.append(row)


x = graph[0]
y =graph[1]
```

```python

year = x
deaths = y
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(year, deaths, color ='maroon', 
        width = 0.4)
 
plt.xlabel("Year")
plt.ylabel("Under -45 deaths in first 43 weeks (England and Wales)")
plt.title("UK Data")
plt.show()
```

```python
x = [] 
y = [] 
graph = []
  
with open('covidandheartattackdeaths.csv','r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    print(plots)
    for row in plots: 
        graph.append(row)


x = graph[0]
y =graph[1]
```

```python
year = x
deaths = y
  
fig = plt.figure(figsize = (10, 5))
# creating the bar plot
plt.bar(year, deaths, color ='maroon', 
        width = 0.4)
plt.set_yticks(np.arange(y.min(), y.max(), 0.005))
plt.xlabel("Year")
plt.ylabel("Under -45 deaths in first 43 weeks (England and Wales)")
plt.title("UK Data")
plt.show()
```
